```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Initialize the models
linear_model = LinearRegression()
lasso_model = Lasso(alpha=0.1)  # Adjust the value of alpha as needed
ridge_model = Ridge(alpha=0.1)  # Adjust the value of alpha as needed
tree_model = DecisionTreeRegressor()

# Fit the models on the training data
linear_model.fit(X_train, y_train)
lasso_model.fit(X_train, y_train)
ridge_model.fit(X_train, y_train)
tree_model.fit(X_train, y_train)

# Make predictions on the testing data
linear_predictions = linear_model.predict(X_test)
lasso_predictions = lasso_model.predict(X_test)
ridge_predictions = ridge_model.predict(X_test)
tree_predictions = tree_model.predict(X_test)

# Calculate R^2 scores for each model
linear_score = r2_score(y_test, linear_predictions)
lasso_score = r2_score(y_test, lasso_predictions)
ridge_score = r2_score(y_test, ridge_predictions)
tree_score = r2_score(y_test, tree_predictions)

print("Linear Regression R^2 Score:", linear_score)
print("Lasso Regression R^2 Score:", lasso_score)
print("Ridge Regression R^2 Score:", ridge_score)
print("Decision Tree Regression R^2 Score:", tree_score)
```

```python
import numpy as np

# Calculate Mean Absolute Percentage Error (MAPE)
def calculate_mape(y_true, y_pred):
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return mape

# Calculate MAPE for each model
linear_mape = calculate_mape(y_test, linear_predictions)
lasso_mape = calculate_mape(y_test, lasso_predictions)
ridge_mape = calculate_mape(y_test, ridge_predictions)
tree_mape = calculate_mape(y_test, tree_predictions)

print("Linear Regression MAPE:", linear_mape)
print("Lasso Regression MAPE:", lasso_mape)
print("Ridge Regression MAPE:", ridge_mape)
print("Decision Tree Regression MAPE:", tree_mape)
```
